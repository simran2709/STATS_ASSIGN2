## THEORY QUESTION

QUES1. What is a random variable in probability theory?
- A random variable is a variable whose value is based on the outcome of a random experiment. It can take different values depending on chance. There are two types: discrete (countable values) and continuous (infinite values).

QUES2. What are the types of random variables?
- There are two main types:
  - Discrete Random Variable – Takes specific, separate values (like 1, 2, 3…).
  - Continuous Random Variable – Takes any value in a range (like 1.1, 1.01, 1.001...).

QUES3. What is the difference between discrete and continuous distributions?
- Discrete distribution deals with countable outcomes (e.g., number of heads in coin toss).
Continuous distribution deals with infinite outcomes within a range (e.g., height, weight).


QUES4. What are probability distribution functions (PDF)?
- PDF is a function that gives the probability of a random variable taking a certain value. For continuous variables, it gives the likelihood density; for discrete, it's called PMF (probability mass function).

QUES5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
- PDF gives probability of exact value.  
CDF gives probability that a value is less than or equal to a certain point.  
CDF is the sum/integral of PDF up to that point.

QUES6. What is a discrete uniform distribution?
- It’s a distribution where each outcome has equal probability. For example, rolling a fair die has equal chance (1/6) for each side.

QUES7. What are the key properties of a Bernoulli distribution?
- 1. Only 2 outcomes: Success (1) or Failure (0)
    - Probability of success = p, failure = 1 - p
    - Mean = p, Variance = p(1 - p)


QUES8. What is the binomial distribution, and how is it used in probability?
- It represents the number of successes in a fixed number of independent trials, with same probability of success in each. Used in scenarios like flipping coins, product testing, etc.

QUES9. What is the Poisson distribution and where is it applied?
- Poisson distribution shows how many times an event happens in a fixed interval of time or space. Used in real-life cases like number of calls at a call center per hour.

QUES10. What is a continuous uniform distribution?
- It represents a continuous range of outcomes where each value within a range is equally likely. Example: random number between 0 and 1.

QUES11. What are the characteristics of a normal distribution?
- Bell-shaped curve
-  Symmetrical around the mean
-  Mean = Median = Mode
-  Follows 68-95-99.7 rule (Empirical Rule)

QUES12. What is the standard normal distribution, and why is it important?
- It’s a normal distribution with mean 0 and standard deviation 1. It's used to simplify calculations by converting values to Z-scores.

QUES13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- CLT states that the sampling distribution of the sample mean approaches normal distribution as the sample size increases, no matter the shape of the original data. It allows us to use normal tools on non-normal data.

QUES14. How does the Central Limit Theorem relate to the normal distribution?
- It helps justify why we use normal distribution for means, even if the original data isn’t normal. When sample size is large, the average of samples becomes approximately normal.

QUES15. What is the application of Z statistics in hypothesis testing?
- Z-statistics helps in determining how far a sample mean is from population mean in standard deviation units. It's used in Z-tests for large samples or known population variance.

QUES16. How do you calculate a Z-score, and what does it represent?
- Z = (X - μ) / σ  
It tells how many standard deviations a data point (X) is from the mean (μ). Positive Z means above average, negative means below.

QUES17. What are point estimates and interval estimates in statistics?
- Point Estimate: A single value used as an estimate of a population parameter (e.g., sample mean).  
Interval Estimate: A range (with confidence) within which the parameter lies (e.g., confidence interval).

QUES18. What is the significance of confidence intervals in statistical analysis?
- Confidence intervals give a range where we expect the true population value to fall. It adds reliability to point estimates, like saying "We are 95% confident that the true mean is between A and B."

QUES19. What is the relationship between a Z-score and a confidence interval?
- Z-scores determine how wide a confidence interval will be. Higher confidence (like 99%) means larger Z and wider interval. For example, 95% CI uses Z = 1.96.

QUES20. How are Z-scores used to compare different distributions?
- Z-scores standardize values from different distributions, letting us compare them on the same scale (mean 0, SD 1).

QUES21. What are the assumptions for applying the Central Limit Theorem?
- Independent samples  
-  Sample size is large (n ≥ 30)  
-  Identical distribution  
-  Finite variance

QUES22. What is the concept of expected value in a probability distribution?
- Expected value is the average value you’d expect after many repetitions. It’s calculated by multiplying each outcome by its probability and adding them up.

QUES23. How does a probability distribution relate to the expected outcome of a random variable?
- Probability distribution shows all possible outcomes and their probabilities. Expected outcome (mean) is derived from the distribution by weighing each value by its chance.

```python
#QUES2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import randint

x = np.arange(1, 7)
pmf = [1/6]*6
plt.stem(x, pmf)
plt.title("Discrete Uniform PMF")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.show()
```

```python
#QUES3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
from scipy.stats import bernoulli

p = 0.6
print("P(X=0):", bernoulli.pmf(0, p))
print("P(X=1):", bernoulli.pmf(1, p))
```

```python
#QUES4. Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram
import numpy as np
import matplotlib.pyplot as plt

data = np.random.binomial(n=10, p=0.5, size=1000)
plt.hist(data, bins=range(12), edgecolor='black')
plt.title("Binomial Distribution Histogram")
plt.show()
```

```python
#QUES5. Create a Poisson distribution and visualize it using Python
from scipy.stats import poisson
import matplotlib.pyplot as plt

x = np.arange(0, 15)
pmf = poisson.pmf(x, mu=4)
plt.stem(x, pmf)
plt.title("Poisson Distribution (mu=4)")
plt.show()

```

```python
#QUES6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
#uniform distribution
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 7)
cdf = np.cumsum([1/6]*6)
plt.step(x, cdf)
plt.title("CDF of Discrete Uniform")
plt.show()

```

```python
#QUES10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
import numpy as np
import matplotlib.pyplot as plt

samples = [np.mean(np.random.exponential(size=50)) for _ in range(1000)]
plt.hist(samples, bins=30, edgecolor='black')
plt.title("CLT with Exponential Distribution")
plt.show()
```

```python
#QUES15. Implement hypothesis testing using Z-statistics for a sample dataset
from scipy.stats import norm

z = (72 - 70) / (5 / (30**0.5))
p = 1 - norm.cdf(z)
print("Z:", z, "P-value:", p)
```

```python
#QUES16. Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
from scipy.stats import norm

data = np.random.normal(50, 10, 100)
mean = np.mean(data)
se = np.std(data)/np.sqrt(len(data))
ci = norm.interval(0.95, loc=mean, scale=se)
print("95% Confidence Interval:", ci)
```

```python
#QUES17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
data = np.random.normal(60, 12, 80)
mean = np.mean(data)
se = np.std(data)/np.sqrt(len(data))
ci = norm.interval(0.95, loc=mean, scale=se)
print("95% CI for mean:", ci)
```
